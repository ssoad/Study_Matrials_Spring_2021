## SSTF

Head Point: 53 <br>
Path Sequence: 53 98 183 37 122 14 124 65 67

```python
from heapq import *
start = int(input("Enter Start Point: "))
queue_str = input("Enter Queue: ")
queue = [int(value) for value in queue_str.split(" ")]
requests = queue.copy()
time = 0
position = start
pre_position = start
n = len(requests)
i=0
heap = []
path = []
while len(requests) > 0:
    for r in requests:
        heappush(heap, (abs(position - r), r))
    x = heappop(heap)[1]
    if x!=start:
      print ("(",x,"-",pre_position,")",end="")
      if len(requests)>1:
        print(" + ",end="")
    else:
      print(end="\n\n")
    time += abs(position - x)
    position = x
    path.append(position)
    pre_position = position
    requests.remove(x)
    heap = []


avg_seek_time = time / n
print("\nPath: ",path,end="\n\n")
print("\nTotal Head Movement of "+str(time)+" Cylinders")
```
