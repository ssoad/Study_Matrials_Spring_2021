## FCFS

Head Point: 53 <br>
Path Sequence: 53 98 183 37 122 14 124 65 67

```python
start = int(input("Enter Start Point: "))
queue_str = input("Enter Queue: ")
queue = [int(value) for value in queue_str.split(" ")]
total = 0
current = start

for i in range(len(queue)):
    total = total + abs(queue[i]-current)
    print(f"({queue[i]}-{current})", end=" ")
    if i!=len(queue)-1:
      print("+", end=" ")
    current = queue[i]

print("\nTotal Head Movement of "+str(total)+" Cylinders")
```
