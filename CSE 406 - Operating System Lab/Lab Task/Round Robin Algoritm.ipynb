## Process Class

```python
class Process:
    def __init__(self, name, time, priority=0):
        self.name = name
        self.time = time
        self.priority = priority

def Average(lst, count=0):
    if count == 0:
      return sum(lst) / len(lst)
    return sum(lst) / count
```

## Round Robin

```python
import operator
procss_list = []
quantam = int(input("Enter Number of Quantam: "))
input_number = int(input("Enter Number of Process: "))
for i in range(input_number):
    name = "P"+str(i+1) #input("Name: ")
    time = float(input("Burst Time: "))
    # priority = int(input("Priority"))
    prcs = Process(name, time)
    prcs.last_time = 0
    prcs.total_time = 0
    procss_list.append(prcs)
new_list = procss_list
init_time = 0
waiting_time=[]
finished =0
count=0
last_time={}
total_time={}
while finished!=len(procss_list):
  for p in procss_list:
    if p.time !=0:
      print(init_time, end=" - ")
      print(p.name, end=" - " )
      if p.last_time!=0:
        p.total_time = (init_time - p.last_time)+p.total_time
      else:
         p.total_time = (init_time - 0)
      if p.time>quantam:
        init_time = init_time + quantam
        p.time = p.time- quantam
      else:
        init_time = init_time + p.time
        p.time = 0
        finished = finished +1
        last_time[p.name]=init_time
      p.last_time = init_time
      count = count + 1 
print(init_time)
waiting_time=[p.total_time for p in procss_list]
average_time = Average(waiting_time,len(procss_list))
# print(waiting_time)
print("Average Waiting Time:", round(average_time,2)," ms")
```
