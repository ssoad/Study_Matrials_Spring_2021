```python
queue_str = input("Enter Queue: ")
queue = [int(value) for value in queue_str.split(" ")]
start = int(input('Give Header Point : '))
queue_2 = queue.copy()
queue_2+=[start,0]
queue_2.sort()
queue_3 = queue_2[0:queue_2.index(start)]
queue_3.append(start)
queue_3 = queue_3[::-1]
queue_4 = queue_2[queue_2.index(start):]
queue_4.sort()
queue_4[queue_4.index(start)]=0
total_cost = []
for i in range(len(queue_3)-1):
    total_cost.append(abs(queue_3[i] - queue_3[i+1]))
    print('({}-{})'.format(queue_3[i],queue_3[i+1]),end=' + ')
    
for i in range(len(queue_4)-1):
    total_cost.append(abs(queue_4[i+1] - queue_4[i]))
    if i == (len(queue_4)-2):
        print('({}-{})'.format(queue_4[i+1],queue_4[i]))
    else:
        print('({}-{})'.format(queue_4[i+1],queue_4[i]),end=' + ')

queue_3.extend(queue_4[1:])
print("Path: ",queue_3)
print('Total Path Cost :',sum(total_cost))
```
