Path : 98 183 37 122 14 124 65 67<br>
Head : 53<br>
End Boundary: 199

```python
queue_str = input("Enter Queue: ")
queue = [int(value) for value in queue_str.split(" ")]
start = int(input('Give Header Point : '))
end_boundary = 199
queue.append(0)
queue.append(start)
queue.append(end_boundary)
queue.sort()
queue_2 = queue
queue_3 = queue_2[:queue_2.index(start)]
queue_4 = queue_2[queue_2.index(start):]
queue_6 = queue_4[:-1]+queue_3[1:]
cost = []

for i in range(len(queue_6)-1):
    cost.append(abs(queue_6[i] - queue_6[i+1]))
    if i == (len(queue_6)-2):
        print('({}-{})'.format(queue_6[i],queue_6[i+1]))
    else:
        print('({}-{})'.format(queue_6[i],queue_6[i+1]),end='+')
print("Path:", queue_6)
print('Total Path Cost :',sum(cost))
```
